```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Crear / Simular el conjunto de datos

Notas:
* Indicar en la variable tot_values cuantos elementos tendremos.




```python
# Establecer la semilla para la reproducibilidad
np.random.seed(0)

tot_values = 200

# Generar valores decimales aleatorios entre 0 y 3 para serieX
serieX = np.random.uniform(0, 3, tot_values)

# Generar valores decimales aleatorios entre 0 y 2 para serieY
serieY = np.random.uniform(0, 2, tot_values)

# Inicializar serieClase con ceros
serieClase = np.zeros(tot_values, dtype=int)

# Modificar serieClase para serieX > 2.3 y serieY > 2
indices_modificar = (serieX > 2.3) & (serieY > 1.0)
serieClase[indices_modificar] = 1

indices_modificar = (serieX < 1) & (serieY > 1.0) & (serieY < 1.50)
serieClase[indices_modificar] = 1

# Crear un DataFrame con las tres series
df = pd.DataFrame({
    'serieX': serieX,
    'serieY': serieY,
    'serieClase': serieClase
})

# Mostrar el DataFrame
print(df)
```

## Gráfica de dispersión

```python
# Gráfica de dispersión

var_x = 'serieX'
var_y = 'serieY'
clase1 = df[df['serieClase']==1]
clase0 = df[df['serieClase']==0]

plt.figure(figsize=(5,5))
plt.xlabel(var_x, fontsize= 10.0)
plt.ylabel(var_y, fontsize=10.0)
plt.scatter(clase0[var_x],
                clase0[var_y],
                label="Clase: 0",
                marker='.',
                c="skyblue",
                s=200
                )
plt.scatter(clase1[var_x],
                clase1[var_y],
                label="Clase: 1",
                marker='.',
                c="red",
                s=200
                )
#plt.legend(bbox_to_anchor=(1, 0.15))

# Líneas adicionales en el gráfico
## código de ayuda para pintar las líneas.
##plt.axvline(x=2, color='gray', linestyle='--')  # Línea vertical en x=2
##plt.axhline(y=1.3, color='gray', linestyle='--')  # Línea horizontal en y=1.3
##plt.xlines(x=2, ymin=0.5, ymax=0.75, color='gray', linestyle='--')  # Línea vertical en x=2, entre y=1 y y=1.5
##plt.hlines(y=0.2, xmin=0.5, xmax=0.6, color='gray', linestyle='--')  # Línea horizontal en y=0.2, entre x=1.3 y x=1.4


plt.axhline(y=1, color='gray', linestyle='--')
plt.vlines(x=1, ymin=1, ymax=2, color='gray', linestyle='--')
plt.hlines(y=1.5, xmin=0, xmax=1, color='gray', linestyle='--')
plt.vlines(x=2.3, ymin=1, ymax=2, color='gray', linestyle='--')

plt.show()
```
